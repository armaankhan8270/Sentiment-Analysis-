```python
df=pd.read_csv('Musical_instruments_reviews.csv',nrows=1000)
```

```python
df=df[['helpful', 'reviewText',
       'overall', 'summary']]
```

```python
def sentiment(raw):
    rating=raw
    if rating>=4:
        return "postive"
    
    else:
        return 'negitive'
    
```

```python
df['sentiment']=df['overall'].apply(lambda x:sentiment(x))
```

```python
df.dropna(inplace=True)
```

```python
df['reviewText']=df['reviewText']+df['summary']
```

```python
df.drop(columns=['summary'],inplace=True)
```

```python
df['reviewText']=df['reviewText'].str.lower()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('wordnet')
```

```python
stopwords_list=stopwords.words('english')
```

```python
import string
ex=string.punctuation
```

```python
def PreProcessing(text):
    PreProcessText=[]
    
    for word in text.split():
        if word not in stopwords_list :
            PreProcessText.append(word)
    return " ".join(PreProcessText)       
```

```python
import re
def review_cleaning(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    PreProcessText=[]
    
    for word in text.split():
        if word not in stopwords_list :
            PreProcessText.append(word)
    return " ".join(PreProcessText)  
```

```python
# df['reviewText']=df['reviewText'].apply(lambda x:PreProcessing(x))
```

```python
# and word not in ex
df['reviewText']=df['reviewText'].apply(lambda x:review_cleaning(x))
```

```python
import ast
def extracPostivetList(text):
    L=ast.literal_eval(text)
    P=L[0]
    
    return P
def extracNegitiveList(text):
    L=ast.literal_eval(text)
   
    N=L[1]
    return N
    
df['Positive']=df['helpful'].apply(lambda x:extracPostivetList(x))
```

```python
df['Negitive']=df['helpful'].apply(lambda x:extracNegitiveList(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=300,ngram_range=(1,2))
x=df['reviewText']

x=tfidf.fit_transform(x).toarray()
     
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
LR=LabelEncoder()
Y=LR.fit_transform(df['sentiment'])
Y=pd.DataFrame(Y)

df['sentiment'].value_counts()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
x= imputer.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,Y,test_size=0.2)
```

```python


rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```
